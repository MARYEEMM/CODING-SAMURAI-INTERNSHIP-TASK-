```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
iris="/content/Iris.csv"
y = ["sepal length","sepal width","petal length","petal width" , "class"]
```

```python
data =pd.read_csv(iris,names=y)
```

```python
data
```

```python
data.head(5)
```

```python
data = sns.load_dataset('iris').rename(columns={'species': 'class'})

# Diagrama de pares coloreado por la clase
sns.pairplot(data, hue="class")

plt.show()
```

```python
x=data.drop("class" ,axis=1)
x
```

```python
y=data["class"]
```

```python
y
```

```python
KNN=KNeighborsClassifier(n_neighbors=3)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
KNN.fit(x_train,y_train)
```

```python
y_pred=KNN.predict(x_test)
```

```python
print("accuracy ", accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
